```python
import h5py
import matplotlib.pyplot as plt
import numpy as np
import sys
import datetime as dt
from xml.etree.ElementTree import Element, SubElement, Comment
from xml.etree.ElementTree import ElementTree, tostring
from xml.etree import ElementTree
from xml.dom import minidom

sys.path.append("/data/laue34/LaueIndex_XML_python/Example/jupyter_files/py_func/")

import fn_LaueXMLfile as LXML
#%pylab inline
#%matplotlib widget
```

```python
# create a dictionary 
def fn_Libs(file_peaks, geo_file, cryst_file):


    extraLibStep = {}
    extraLibDet = {}
    extraLibPeaksXY = {}
    generated_on = str(dt.datetime.now())

    extraLibStep['title'] = '2023-1'
    extraLibStep['sampleName'] = 'Ni2_10mN_test_Polaris'
    extraLibStep['userName']  =  'dsheyfer'
    extraLibStep['beamline']  = '34ID-E'

    with open(file_peaks,encoding='windows-1252', errors='ignore') as f:
        lines = f.readlines()
    for ii in range(len(lines)):
            if lines[ii][0:10] == '$scanNum\t\t':
                extraLibStep['scanNum'] = str(lines[ii][10:16])
            if lines[ii][0:10] == '$beamBad\t\t':
                extraLibStep['beamBad'] = str(lines[ii][10:11])
            if lines[ii][0:10] == '$Xsample\t\t':
                s = lines[ii]
                s = s.replace("\n", "")
                s = s.split('\t')
                #print(s)
                extraLibStep['Xsample'] = str(s[2])
            if lines[ii][0:10] == '$Ysample\t\t':
                s = lines[ii]
                s = s.replace("\n", "")
                s = s.split('\t')
                #print(s)
                extraLibStep['Ysample'] = str(s[2])
            if lines[ii][0:10] == '$Zsample\t\t':
                s = lines[ii]
                s = s.replace("\n", "")
                s = s.split('\t')
                #print(s)
                extraLibStep['Zsample'] = str(s[2])
            if lines[ii][0:8] == '$depth\t\t':
                s = lines[ii]
                s = s.replace("\n", "")
                s = s.split('\t')
                #print(s)
                extraLibStep['depth'] = str(s[3])
            if lines[ii][0:9] == '$energy\t\t':
                s = lines[ii]
                s = s.replace("\n", "")
                s = s.split('\t')
                #print(s)
                extraLibStep['energy'] = str(s[3])
            if lines[ii][0:18] == '$hutchTemperature\t':
                s = lines[ii]
                s = s.replace("\n", "")
                s = s.split('\t')
                #print(s)
                extraLibStep['sampleDistance'] = str(s[1])
            if lines[ii][0:16] == '$sampleDistance\t':
                s = lines[ii]
                s = s.replace("\n", "")
                s = s.split('\t')
                #print(s)
                extraLibStep['sampleDistance'] = str(s[1])

            if lines[ii][0:11] == '$exposure\t\t':
                s = lines[ii]
                s = s.replace("\n", "")
                s = s.split('\t')
                #print(s)
                extraLibDet['exposure'] = str(s[2])

            if lines[ii][0:11] == '$totalSum\t\t':
                s = lines[ii]
                s = s.replace("\n", "")
                s = s.split('\t')
                #print(s)
                extraLibDet['totalSum'] = str(s[2]) 

            if lines[ii][0:19] == '$sumAboveThreshold\t':
                s = lines[ii]
                s = s.replace("\n", "")
                s = s.split('\t')
                #print(s)
                extraLibDet['sumAboveThreshold'] = str(s[1]) 

            if lines[ii][0:19] == '$numAboveThreshold\t':
                s = lines[ii]
                s = s.replace("\n", "")
                s = s.split('\t')
                #print(s)
                extraLibDet['numAboveThreshold'] = str(s[1]) 




    extraLibStep['date'] = str(generated_on)
    extraLibStep['CCDshutter'] = 'out'
    extraLibStep['lightOn'] = '0'
    extraLibStep['monoMode'] = 'white slitted'




    
    extraLibDet['detectorID'] = 'PE1621 723-3335' # depends on the used detector
    extraLibDet['Nx'] = '2048'
    extraLibDet['Ny'] = '2048'
    
    extraLibDet['geoFile'] = geo_file

    
    
###################### I have to put here manualy still; depends on the crystal structure; need some kinda xml structure reader from the lattice structure files

    xtl_str       = Element('xtl')
    structureDesc_       = SubElement(xtl_str, 'structureDesc')
    structureDesc_.text = 'Ni'
    xtlFile_       = SubElement(xtl_str, 'xtlFile')
    xtlFile_.text = cryst_file
    SpaceGroup_       = SubElement(xtl_str, 'SpaceGroup')
    SpaceGroup_.text = '225'
    latticeParameters_       = SubElement(xtl_str, 'latticeParameters')
    latticeParameters_.set('unit', 'nm')
    latticeParameters_.text = '0.35238, 0.35238, 0.35238, 90, 90, 90'
    atom_       = SubElement(xtl_str, 'atom')
    atom_.set('n', '1')
    atom_.set('symbol', 'Ni')
    atom_.set('Ni', '30')
    atom_.set('label', 'Ni001')
    atom_.text = '0 0 0'
    

#     xtl_CdTe       = Element('xtl')
#     structureDesc_       = SubElement(xtl_CdTe, 'structureDesc')
#     structureDesc_.text = 'CdTe'
#     xtlFile_       = SubElement(xtl_CdTe, 'xtlFile')
#     xtlfile = '/private/nfs/data/xmd34/2022-3/XEOL1122/post_analysis/Laue_analysis/HASshort_laue/CdTe.xml'
#     xtlFile_.text = xtlfile
#     SpaceGroup_       = SubElement(xtl_CdTe, 'SpaceGroup')
#     SpaceGroup_.text = '225'
#     latticeParameters_       = SubElement(xtl_CdTe, 'latticeParameters')
#     latticeParameters_.set('unit', 'nm')
#     latticeParameters_.text = '0.648 0.648 0.648 90 90 90'
#     atom_       = SubElement(xtl_CdTe, 'atom')
#     atom_.set('n', '1')
#     atom_.set('symbol', 'Cd')
#     atom_.set('Z', '48')
#     atom_.set('label', 'Cd0')
#     atom_.text = '0 0 0'
    
#     atom2_       = SubElement(xtl_CdTe, 'atom')
#     atom2_.set('n', '2')
#     atom2_.set('symbol', 'Te')
#     atom2_.set('Z', '52')
#     atom2_.set('label', 'Te4')
#     atom2_.text = '0.25 0.25 0.25'



#     xtl_Zn       = Element('xtl')
#     structureDesc_       = SubElement(xtl_Zn, 'structureDesc')
#     structureDesc_.text = 'Zinc'
#     xtlFile_       = SubElement(xtl_Zn, 'xtlFile')
#     xtlfile = '/private/nfs/data34a/Run2022-3/Kent1022/ZnCu111_1um/Zn.xml'
#     xtlFile_.text = xtlfile
#     SpaceGroup_       = SubElement(xtl_Zn, 'SpaceGroup')
#     SpaceGroup_.text = '194'
#     latticeParameters_       = SubElement(xtl_Zn, 'latticeParameters')
#     latticeParameters_.set('unit', 'nm')
#     latticeParameters_.text = '0.26649, 0.26649, 0.49468, 90, 90, 120'
#     atom_       = SubElement(xtl_Zn, 'atom')
#     atom_.set('n', '1')
#     atom_.set('symbol', 'Zn')
#     atom_.set('Z', '30')
#     atom_.set('label', 'Zn1')
#     atom_.text = '0.33333 0.66667 0.25'


    return extraLibStep, extraLibDet, xtl_str#, xtl_Zn
```

```python
# set_file_range

scanPoint = np.arange(1,3,1)
depthRange = np.arange(100, 102, 1)
#depthRange = np.nan #if no depthRange

#set data folder for reading
filefolder = '/data/laue34/repacks_10mn/consgeo/'
filename_ = 'out_'
filename_prefix = ''

#set data folder for saving results
geo_file = '/data/laue34/LaueIndex_XML_python/Example/jupyter_files/geoN_2023-02-07_19-19-10.xml'
cryst_file = '/data/laue34/LaueIndex_XML_python/Example/jupyter_files/Ni.xml'

#generate the list of files name for analysis
fnames = []

if np.isnan(depthRange).any():
    for ii in range(len(scanPoint)):
        # if the depthRange exist
        fnames.append(filename_prefix + filename_ + str(scanPoint[ii]))
else:
    for ii in range(len(scanPoint)):
        for jj in range (len(depthRange)):
            fnames.append(filename_prefix + filename_ + str(scanPoint[ii]) + '_' + str(depthRange[jj]))

print(fnames)    



# indexed results

filefolderIndex = '/data/laue34/LaueIndex_XML_python/Example/jupyter_files/results/out_index1/'
index_prefix = 'index_'
p2q_prefix = 'p2q_'
peaks_prefix = 'peaks_'


xml_file_out = '/data/laue34/LaueIndex_XML_python/Example/jupyter_files/results/xml_out_index1.xml'

```

```python

generated_on = str(dt.datetime.now())

Allsteps_ = Element('AllSteps')

for ii in range(len(fnames)):

    file_peaks = filefolderIndex + peaks_prefix + fnames[ii] + '.txt'
    file_p2q = filefolderIndex + p2q_prefix + fnames[ii] + '.txt'
    file_index = filefolderIndex + index_prefix + fnames[ii] + '.txt'
    
    [extraLibStep, extraLibDet, xtl_CdTe] = fn_Libs(file_peaks, geo_file, cryst_file)
    
    filefull = filefolder + fnames[ii] + '.h5'
    extraLibDet['inputImage'] = filefull
    step1 = LXML.fn_createStepXML(file_peaks, file_p2q, file_index, extraLibStep, extraLibDet, xtl_CdTe)

    Allsteps_.append(step1)


rough_string = ElementTree.tostring(Allsteps_, 'utf-8')
reparsed = minidom.parseString(rough_string)
to_file = reparsed.toprettyxml(indent="    ")

#print(test)

with open(xml_file_out, "w") as f:
    f.write(to_file)
    
```
